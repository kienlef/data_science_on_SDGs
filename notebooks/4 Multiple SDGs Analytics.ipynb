# Goal is to align all data sets to a normalized key.

 A key is a unique identfier
 For the current purpose we would like to aling on the country code and year.
 
 * read a basic file and extract the unique country codes
 * generate a complete list of years, e.g. 1980 to today
 * ensur that each country has all years 
 * the resulting base index should result in nr_of_rows = nr_of_countries * nr_of_years


```python
## get an inital list of country codes
df_data=pd.read_csv('8.1.1 gdp-per-capita-worldbank.csv',)
df_data.head()
```

```python
## get the unique code list of the world >180 countries
unique_code=pd.DataFrame({'Code':df_data.Code.unique()})
unique_code=unique_code.dropna()
unique_code
```

```python
# generate a year list
all_years=pd.DataFrame({'Year':range(1980,2021,1)})
#all_years
```

```python
# start to blow up the full index list
full_index=pd.DataFrame(columns=['Year', 'Code'])

for each_code in unique_code['Code']:
    ''' walk through all countries and append the harmonized year range'''
    full_index = full_index.append(pd.DataFrame({'Year':all_years.Year, 'Code': each_code} ))
full_index
```

# Loop through all files you would like to inspect and build the Matrix

```python
for each in files:
    df_data=pd.read_csv(each,)
    #print(each)
    try:
        ## rename column to get a link to the goal /index number
        col_rename=df_data.columns
        df_data.rename({col_rename[3]:each.strip('.csv')},axis='columns',inplace=True)
        
        # drop entity name otherwise multiple copies of the same information
        df_data=df_data.drop(['Entity'],axis=1)
        full_index=pd.merge(full_index,df_data,how='left',on=['Year','Code'])
        print(each+'......done')
    except:
        print(each+'......failed')
        pass
    
```

# Correlation analysis of different indicators

https://en.wikipedia.org/wiki/Correlation
In statistics, correlation or dependence is any statistical relationship, whether causal or not, between two random variables or bivariate data. In the broadest sense correlation is any statistical association, though it actually refers to the degree to which a pair of variables are linearly related.

### It gives a linear dependency 
### !!!!!!! correlation does not imply causation !!!!!!

```python
my_country_corr=my_country.corr()
```

```python
my_country_corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

sns.set_theme()
sns.color_palette("magma", as_cmap=True)

# mask out the upper triangular for drawing
mask = np.triu(np.ones_like(my_country_corr, dtype=bool)) # only look at the diagonals

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(16, 12))

# Draw a heatmap with the numeric values in each cell

sns.heatmap(my_country_corr,
            mask=mask,
            annot=True,  
            linewidths=.5, 
            square=True,
            ax=ax)


#plt.savefig('save_USA_correlation.png', dpi=300)
```
